```python
from roboflow import Roboflow
rf = Roboflow(api_key="x6TWMeDpXTqAu2duUUOY")
project = rf.workspace().project("thermal-dogs-and-people")
model = project.version(3).model

# infer on a local image
print(model.predict("original.jpg", confidence=40, overlap=30).json())

# visualize your prediction
model.predict("/content/original.jpg", confidence=40, overlap=30).save("prediction.jpg")

# infer on an image hosted elsewhere
# print(model.predict("URL_OF_YOUR_IMAGE", hosted=True, confidence=40, overlap=30).json())
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Load the original and prediction images
original_image = Image.open("/content/original.jpg")
prediction_image = Image.open("/content/prediction.jpg")

# Create a subplot with two columns
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Display the original image in the first column
axes[0].imshow(original_image)
axes[0].set_title("Original Image")

# Display the prediction image in the second column
axes[1].imshow(prediction_image)
axes[1].set_title("Prediction Image")

# Remove axis ticks and labels for a cleaner display
for ax in axes:
    ax.axis("off")

# Show the images side by side
plt.show()
```
